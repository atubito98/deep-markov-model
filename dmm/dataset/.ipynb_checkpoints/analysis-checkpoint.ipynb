import numpy as np

from sklearn.cross_decomposition import CCA

# --- 1. Create a Toy Dataset ---
# Let's simulate two sessions with some shared underlying dynamics.
n_time = 256
channels= 96
n_shared_patterns = 5
k_components = 15 # Number of components for PCA/CCA

# Create some smooth, underlying patterns (the "themes")
np.random.seed(42)
shared_dynamics = np.random.randn(n_time, n_shared_patterns)
for i in range(n_shared_patterns):
    shared_dynamics[:, i] = np.convolve(shared_dynamics[:, i], np.ones(30)/30, mode='same')

# --- Session A ---
# Neurons in session A are a linear combination of these patterns + noise
weights_A = np.random.randn(n_shared_patterns, channels)
noise_A = np.random.randn(n_time, channels) * 0.3
X_A = shared_dynamics @ weights_A + noise_A

# --- Session B (Consistent Session) ---
# Neurons in session B are a DIFFERENT combination of the SAME patterns + noise
# This simulates recording different neurons that are part of the same circuit.
weights_B_consistent = np.random.randn(n_shared_patterns, channels)
noise_B = np.random.randn(n_time, channels) * 0.3
X_B_consistent = shared_dynamics @ weights_B_consistent + noise_B

# --- Session C (Inconsistent Session) ---
# This session has completely different underlying dynamics.
other_dynamics = np.random.randn(n_time, n_shared_patterns)
weights_C = np.random.randn(n_shared_patterns, channels)
X_C_inconsistent = other_dynamics @ weights_C + noise_B

# --- 2. Step-by-Step Analysis ---

def analyze_session_consistency(session1_data, session2_data, n_components):
    """Performs PCA and CCA to compare two sessions."""
    
    # Step 2a: Center the data
    session1_data -= session1_data.mean(axis=0)
    session2_data -= session2_data.mean(axis=0)
    
    # Step 2b: Run PCA on each session
    pca1 = PCA(n_components=n_components)
    pca2 = PCA(n_components=n_components)
    
    principal_components1 = pca1.fit_transform(session1_data)
    principal_components2 = pca2.fit_transform(session2_data)
    
    # Step 3: Run CCA to align the two PC subspaces
    cca = CCA(n_components=n_components)
    cca.fit(principal_components1, principal_components2)
    
    # The `transform` method gives you the canonical variates, but we need the correlations.
    # A simple way to get the correlation scores is to transform and correlate.
    X_c, Y_c = cca.transform(principal_components1, principal_components2)
    
    # Calculate correlations along the diagonal
    corrs = np.array([np.corrcoef(X_c[:, i], Y_c[:, i])[0, 1] for i in range(n_components)])
    
    return corrs

# --- 4. Run and Interpret ---
consistent_corrs = analyze_session_consistency(X_A, X_B_consistent, k_components)
inconsistent_corrs = analyze_session_consistency(X_A, X_C_inconsistent, k_components)

print(f"Correlations (Session A vs B - Consistent):")
print(np.round(consistent_corrs, 2))
# Expected output: High correlations for the first ~5 components, then drop-off

print(f"\nCorrelations (Session A vs C - Inconsistent):")
print(np.round(inconsistent_corrs, 2))
# Expected output: Low correlations across the board

# Plotting the results
plt.figure(figsize=(10, 5))
plt.plot(range(1, k_components + 1), consistent_corrs, 'o-', label='Consistent Sessions (A vs B)')
plt.plot(range(1, k_components + 1), inconsistent_corrs, 'o-', label='Inconsistent Sessions (A vs C)')
plt.title('Cross-Session Consistency via CCA')
plt.xlabel('Canonical Component Index')
plt.ylabel('Correlation Coefficient')
plt.xticks(range(1, k_components + 1))
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()
plt.show()

def channel2grid(data):
    n=len(data.shape)
    out_shape = [10, 10]
    out_shape = list(data.shape[:-1]) + out_shape
    grid_data = np.zeros(out_shape)
    
    valid_channels = {k: v for k, v in ch2grid.items() if k < data.shape[-1]}
    for ch, (i, j) in valid_channels.items():
        if n == 1:
            grid_data[i, j] = data[ch]
        elif n == 2:
            grid_data[:, i, j] = data[:, ch]
        elif n == 3:
            grid_data[:, :, i, j] = data[:, :, ch]
        else:  # n == 4
            grid_data[:, :, :, i, j] = data[:, :, :, ch]
    return grid_data

```python
from sklearn.decomposition import PCA
import os

def download_PCA_data(data_path_array, selection, n_pca):
    d = len(data_path_array)
    all_data = []
    all_name = []
    for j in range(d):
        (data,
        RT,
        SSD,
        direction,
        session,
        subject) = load_test_set(list([data_path_array[j]]),selection)
        
        name = os.path.splitext(os.path.basename(data_path_array[j]))[0]
        all_name.append(name)
        data_postGO = data[:, 56:]
        s, steps, _ = data_postGO.shape
        X = data_postGO.reshape(-1, data_postGO.shape[2])
        pca = PCA(n_components=n_pca, random_state=42)
        X_pca = pca.fit_transform(X)
        #print(X_pca.shape)
        data_pca = X_pca.reshape(s, steps, n_pca)
        all_data.append(data_pca)

    return np.array(all_data), np.array(all_name)
```

```python
from sklearn.decomposition import PCA

def download_PCA_data(data_path_array, selection):
    d = len(data_path_array)
    all_data = []
    for j in range(d):
        (data,
        RT,
        SSD,
        direction,
        session,
        subject) = load_test_set(list([data_path_array[j]]),selection)
        
        data_postGO = data[:, 56:]
        X = data_postGO.reshape(-1, data_postGO.shape[2])
        pca = PCA(n_components=3, random_state=42)
        X_pca = pca.fit_transform(X)
        #print(X_pca.shape)
        all_data.append(X_pca)

    return np.array(all_data)

def run_pairwise_cca(all_data, n_components=5):
    """
    Performs pairwise CCA on a list of session data arrays.
    """
    num_sessions = len(all_data)
    cca_scores = np.zeros((num_sessions, num_sessions))

    # Get the number of features (channels) from the first session
    #n_features = all_data[0].shape[2]

    print(f"\nStarting pairwise CCA for {num_sessions} sessions with {n_components} components...")

    correlations = np.zeros((num_sessions, num_sessions, n_components))
    for i in range(num_sessions):
        for j in range(i, num_sessions):
            if i == j:
                correlations[i, j] = 1.0  # A session is perfectly correlated with itself
                continue

            print(f"  Comparing session {i+1} and {j+1}...")

            # 1. Prepare data: reshape to (samples, features)
            X = all_data[i]#.reshape(-1, n_features)
            Y = all_data[j]#.reshape(-1, n_features)

            # 2. Handle mismatched sample sizes by subsampling the larger dataset
            n_samples_X = X.shape[0]
            n_samples_Y = Y.shape[0]

            if n_samples_X > n_samples_Y:
                indices = np.random.choice(n_samples_X, n_samples_Y, replace=False)
                X = X[indices, :]
            elif n_samples_Y > n_samples_X:
                indices = np.random.choice(n_samples_Y, n_samples_X, replace=False)
                Y = Y[indices, :]
            
            # 3. Initialize and fit CCA
            cca = CCA(n_components=n_components)
            cca.fit(X, Y)

            # 4. Transform data to the canonical space
            X_c, Y_c = cca.transform(X, Y)

            # 5. Calculate correlations for each component and average them
            for k in range(n_components):
                # Calculate the Pearson correlation between the k-th component of X and Y
                correlations[i, j, k] = np.corrcoef(X_c[:, k], Y_c[:, k])[0, 1]
                correlations[j, i, k] = np.corrcoef(X_c[:, k], Y_c[:, k])[0, 1]
                print(f"Correlation for the component {k}: {correlations[i, j, k]}")
            
    cca_scores = np.mean(correlations, axis=2)   
    print("\nCCA calculations complete.")
    return cca_scores, correlations

def plot_cca_heatmap(scores, labels):
    """
    Plots the CCA scores as a heatmap.
    """
    plt.figure(figsize=(14, 12))
    sns.heatmap(
        scores, 
        annot=True, 
        fmt=".2f", 
        cmap='viridis', 
        xticklabels=labels, 
        yticklabels=labels
    )
    plt.title(f'Average Canonical Correlation Across Top {N_COMPONENTS} Components', fontsize=16)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```
